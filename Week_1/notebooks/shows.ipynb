# Netflix Movies & TV Shows Dataset

## Load the dataset into Python

```python
# Definition for getting the data and modules
import pandas as pd
url='https://www.kaggle.com/datasets/shivamb/netflix-shows'
```

```python
# Displaying the first and last rows
delimiters = ['\t', ',', ';',  '|', ' ']
for delim in delimiters:
    try:
        df = pd.read_csv(url, sep=delim)
        print(f"Success with delimiter: '{delim}'")
        print(df.head())
        print(df.tail())
        break
    except Exception as e:
        print(f"Failed with delimiter '{delim}': {e}")
```
