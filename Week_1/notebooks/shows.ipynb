# Netflix Movies & TV Shows Dataset

```python
# Definition for getting the data and modules
import pandas as pd
import kagglehub as ka
import numpy as np

url_path = 'https://www.kaggle.com/datasets/shivamb/netflix-shows'
data_path = "../data/netflix_titles.csv"
```

```python
# Download the dataset directly to the 'data' directory
path = ka.dataset_download("shivamb/netflix-shows", output_dir="../data")
```

```python
# Displaying the first and last rows
df = pd.read_csv(data_path)
print(df.head(5))
print(df.tail(5))
```

```python
# Number of rows and columns
rows, columns = df.shape
print(f"Rows: {rows}, Columns:{columns}")
```

```python
# Data types of all columns
print(df.dtypes)
```

```python
# Numerical features
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()

print(list(df.select_dtypes(include=np.number).columns))
```

```python
# Categorical features
cat_df = df.select_dtypes(include=['object', 'category'])
print(cat_df)
```

```python
# Possible unique identifiers (primary keys)
potential_keys = []
for col in df.columns:
    if df[col].nunique() == len(df):
        potential_keys.append(col)

print(f"Potential primary keys: {potential_keys}")
```

```python
def describe_dataset(df):
    """Generate a brief description of the dataset"""
    
    # Calculate column types properly
    numeric_cols = df.select_dtypes(include=['number']).columns
    text_cols = df.select_dtypes(include=['object', 'string']).columns  # Both old and new text types
    categorical_cols = df.select_dtypes(include=['category']).columns
    datetime_cols = df.select_dtypes(include=['datetime64']).columns
    bool_cols = df.select_dtypes(include=['bool']).columns
    
    # Count each type
    n_numeric = len(numeric_cols)
    n_text = len(text_cols)
    n_categorical = len(categorical_cols)
    n_datetime = len(datetime_cols)
    n_bool = len(bool_cols)
    
    # Calculate 'other' (everything not explicitly categorized)
    n_other = len(df.columns) - n_numeric - n_text - n_categorical - n_datetime - n_bool
    
    # Get columns with missing values (excluding those with 0 missing)
    cols_with_missing = df.columns[df.isnull().any()].tolist()
    
    # Build the description
    description = f"""
    DATASET OVERVIEW
    ----------------
    • Rows: {df.shape[0]:,} records
    • Columns: {df.shape[1]} features
    • Memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB
    
    COLUMN TYPES:
    • Numeric: {n_numeric} columns {f'- {list(numeric_cols)[:5]}...' if n_numeric > 5 else f'- {list(numeric_cols)}' if n_numeric > 0 else ''}
    • Text (string/object): {n_text} columns {f'- {list(text_cols)[:5]}...' if n_text > 5 else f'- {list(text_cols)}' if n_text > 0 else ''}
    • Categorical: {n_categorical} columns {f'- {list(categorical_cols)}' if n_categorical > 0 else ''}
    • DateTime: {n_datetime} columns {f'- {list(datetime_cols)}' if n_datetime > 0 else ''}
    • Boolean: {n_bool} columns {f'- {list(bool_cols)}' if n_bool > 0 else ''}
    • Other: {n_other} columns {f'- {list(df.columns[~df.columns.isin(list(numeric_cols) + list(text_cols) + list(categorical_cols) + list(datetime_cols) + list(bool_cols))])}' if n_other > 0 else ''}
    
    MISSING VALUES:
    • Total missing: {df.isnull().sum().sum():,} cells ({df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100:.1f}% of data)
    • Columns with missing: {cols_with_missing if cols_with_missing else 'None'}
    
    SAMPLE DATA (First 3 rows):
    {df.head(3).to_string()}
    """
    
    return description

# Usage
print(describe_dataset(df))
```
